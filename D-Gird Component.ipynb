```python
a,b = [int(i) for i in input().split()]
k = 100

ans_b = [['#' * 100] for _ in range(50)]
ans_w = [['.' * 100] for _ in range(50)]

rest_a = a-1
rest_b = b-1

a_loop = rest_a // 50
a_last = rest_a - a_loop
b_loop = rest_b // 50
b_last = rest_b - b_loop

for _ in range(a_loop+1):
    if _ == a_loop:
        ans_b[2*_] = '#.'*a_last + '#' * (100 - 2*a_last)
        
    else:
        ans_b[2*_] = '#.'*50
        
for _ in range(b_loop+1):
    if _ == b_loop:
        ans_w[2*_+1] = '#.'*a_last + '.' * (100 - 2*a_last)
        
    else:
        ans_w[2*_+1] = '#.'*50
    
for i in range(50):
    print(ans_b[i])
    
for i in range(50):
    print(ans_w[i])
```
